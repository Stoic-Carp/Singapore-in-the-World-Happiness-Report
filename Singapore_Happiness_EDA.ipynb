## This is a notebook to explore the happiness index of Singapore in relation to the rest of the world. 

Data is from Kaggle: https://www.kaggle.com/ajaypalsinghlo/world-happiness-report-2021 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading in data
current_df = pd.read_csv('data/world-happiness-report-2021.csv')
past_df = pd.read_csv('data/world-happiness-report.csv')
```

```python
current_df.describe()
```

```python
current_df.sort_values(by='Ladder score')
top_hap = current_df[['Country name', 'Ladder score']][:10]
bottom_hap = current_df[['Country name', 'Ladder score']][-10:]
sg_hap = current_df[current_df['Country name'] == 'Singapore']
sg_hap = sg_hap[['Country name', 'Ladder score']]
top_bottom_hap = top_hap.append(bottom_hap)
top_bottom_sg_hap = top_bottom_hap.append(sg_hap).sort_values(by='Ladder score', ascending=False)
top_bottom_sg_hap
```

```python
sns.set_style('white')

sns.barplot(x='Ladder score', y='Country name', data=top_bottom_sg_hap, palette='coolwarm')
```
